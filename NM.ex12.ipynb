```python
%load_ext sympy.interactive.ipythonprinting
%pylab inline

from __future__ import division
import sympy as sym
from sympy import Symbol, exp
from numpy import *
```

Linear Fredholm integral equation:

$$y(x) = \int_{a}^{b} k(t, x) y(t)dt + f(x) $$

```python
x = Symbol('x')
f = sym.exp(-x);
k = sym.exp(x);
d = 2
```

The next step is to find function's formula in operator domain and then transforming it into time domain. Thanks to Python SymPy library, process of finding Laplace transforms and computing function's formula in time domain can be automated.

```python
from sympy.integrals import laplace_transform
from sympy.abc import s
F = laplace_transform(f,x,s)
K = laplace_transform(k,x,s)
Y = F[0]/(1-d*K[0])
Y
```

```python
from sympy.integrals import inverse_laplace_transform
y = inverse_laplace_transform(Y,s,x)
y
```

After transformations function's formula becomes $$ y = \frac{1}{2} ( e^{3x} + e^{-x} )$$ for $x \ge 0$

After finding analytic formula of our solution it is good to see the shape of a function. But before that it is recomended to transform analytical formula into numerical array and use it further.

```python
plt.rc('figure', figsize=(7, 6))
#border of domain
a = 0
b = 1
# number of regions
N = 1000
# step size
h = (b - a)/N
#calculating argument values and functions values
x_vec = numpy.arange(a,b+h,h) # start, stop,  step
y_vec = numpy.array([y.subs(x,xx).evalf(6) for xx in x_vec])


fig, ax = subplots()
ax.plot(x_vec, y_vec, label = '$0.5 \cdot( e^{3x} + e^{-x} )$');
grid()
ax.set_xlabel('$x$', fontsize='15')
ax.set_ylabel('$y$', fontsize='15')
ax.legend(loc=2, fontsize='15');
```

From now on I can start numerical search of function values. The method used is the Neumann's iterative method which was presented during laboratory excercises. But first let's create function resposible for numerical integration (rectangular method).

```python
def rect(y,x,h):
    #taking vectors x and y
    #assume constant step size
    int = 0
    if x.size == 1:
        return y[0]*h
    else:
        for ind in range(x[:-1].size):
            int += y[ind]*h
    return int
```

Then one can create implementation of kernel and free term functions. In my case such definitions seems redundant because of simplicity of functions, but using that methods allows easy change to more complex.

$k(x) = e^x, f(x) = e^{-x} $

```python
def k_n(x):
    return exp(x)

def f_n(x):
    return exp(-x)
```

# Implementation in Python

Next few lines present the main part of a Neumann method. The values are calculated according to formula:

$$ y_i(x) = \lambda \int^x_0 k(x - t)y_{i-1}(t) dt + f(x)$$
Which of the main part of equation is the convolution. The problem is that intergration cannot be computed due to discretization and a numerical integration method has to be chosen. In laboratory excercises, the simplest one - rectalgular method was suggested. After transformation into discrete form and choosing rectangular method as integration method, the formula for function's values at point $x_j$ is such:
$$ y_i(x_j) = \lambda \sum^j_{k=1} k(x_j - x_k)y_{i-1}(x_k)h + f(x_j)$$

To calculate functions values in next iteration one has to perform two loops:

* one connected with integration
* second one connected with calculating function's values at succesive points

In order to compute many iterations another loop is needed

```python
def Neumann_rect(x_vec, it_num, a, k_n, f_n):
    """Neumann method of solving integral equations.
    Using rectangular method as integration method

    Keyword arguments:
    x_vec -- vector of arguments
    it_num -- number of iterations
    a -- initial value of 0th iteration
    k_n -- kernel function
    f_n - free term function

    """
    #initial condition
    y_it = a*np.ones((it_num+1,x_vec.size))
    #outside loop
    for n in range(1,it_num+1):
        y_i = np.zeros(x_vec.size)
        #internal loops which iterates through every point in x_vec
        for j in range(x_vec.size):
            y_i[j] = (d * rect(k_n(x_vec[j]-x_vec[:j])*y_it[n-1][:j],x_vec[:j], h) + f_n(x_vec[j]))
        y_it[n] = y_i
    return y_it
```

## Different integration method

It is also very interesting to see the influence of integration method on the result. First the trapezoid method was used. For that purpose algorithm's code had to be changed. rect() function was changed to trapz() method from scipy.integrate library

```python
from scipy.integrate import trapz,simps
def Neumann_trapz(x_vec, it_num, a, k_n, f_n):
    """Neumann method of solving integral equations.
    Using scipy.integrate.trapz method as integration method

    Keyword arguments:
    x_vec -- vector of arguments
    it_num -- number of iterations
    a -- initial value of 0th iteration
    k_n -- kernel function
    f_n - free term function

    """
    #initial condition
    y_it = a*np.ones((it_num+1,x_vec.size))
    #outside loop
    for n in range(1,it_num+1):
        y_i = np.zeros(x_vec.size)
        #internal loops which iterates through every point in x_vec
        for j in range(x_vec.size):
            # CHANGE! Instead of rect(x,y,h), it's used trapz(x,y)
            y_i[j] = d * trapz( k_n( x_vec[j]-x_vec[:j]) * y_it[n-1][:j],x_vec[:j]) + f_n( x_vec[j]);
        y_it[n] = y_i;
    return y_it
```

```python
y_trap = Neumann_trapz(x_vec, 10, 0, k_n, f_n)
```

```python
title('Comparison between different integration \nmethods in Neumann algorithm in $5^{th}$ iteration')
plot(x_vec,y_vec,'black', ls='-',label = 'True values')
plot(x_vec,y_trap[5],'r',ls='-.', label = 'Trapezoid')
plot(x_vec,y_it[5],'b', ls=':',label = 'Rectangular')
grid()
xlim([0,1])
legend(loc=2);
```

The result doesn't differ at all. But one has to remember that sample size is $N=1000,$   which a lot for assumed domain.
